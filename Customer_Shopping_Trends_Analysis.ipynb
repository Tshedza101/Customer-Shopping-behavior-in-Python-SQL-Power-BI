```python
import pandas as pd
file_path = r'C:\Users\Tshedza\Documents\data-analysis-SQL-Python-PowerBI\customer_shopping_behavior.csv'
df = pd.read_csv(file_path)
```

```python
#check missing values or nulls
df.isnull().sum()
```

```python
#clean null values - fill missing values in the review rating column with the median rating of the product category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isnull().sum() #cleaned
```

```python
#clean column names, remove spaces and use snake_casing
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column 'age_group',
labels = ['Young Adult', ' Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels) 
#qcut splits ages into 4 age categories
```

```python
df[['age','age_group']].head(10)
```

```python
# create a coulumn 'purchase_frequency_days'

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
#take necessary data(discount and promo code contains same data)
df[['discount_applied','promo_code_used']].head(10)
```

```python
#proof
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
#drop promo_code_used column
df = df.drop('promo_code_used', axis = 1)
```
